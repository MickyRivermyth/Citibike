### Create a dataframe with only necessary columns

```python
# Select all necessary columns and put them in a list called columns

columns = ['Unnamed: 0',
           'day_of_week',
          'start_hour',
          'start_time',
          'end_time',
          'trip_duration',
          'start_station_id',
          'start_station_name',
          'start_station_latitude',
          'start_station_longitude',
          'end_station_id',
           'end_station_name',
          'end_station_latitude',
          'end_station_longitude',
          'subscriber',
          'birth_year',
          'gender',
          'start_station_coords',
          'end_station_coords',
          'trip_length',
          'start_date',
          'end_date']
```

```python
station_rec = citibike_df[columns]
```

```python
station_rec.head()
```

```python
# Create data frame with jsut the start_station_id and day_of_week
data_to_plot = station_rec[['Unnamed: 0', 'start_station_latitude', 'start_station_longitude', 'start_station_coords', 'day_of_week', 'trip_length']]
data_to_plot.head()
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [40.7, -74], zoom_start = 10)

# Choropleth maps bind Pandas Data Frames and json geometries to allow us a quick visualization of data combination
folium.Choropleth(
    geo_data = ny_geo,
    data = data_to_plot,
    columns = ['Unnamed: 0', 'day_of_week', 'start_station_coords'],
    key_on = 'feature.properties.name',
    fill_color = 'Pastel2', fill_opacity=.6, line_opacity=0.1,
    legend_name = "day_of_week").add_to(map)
folium.LayerControl().add_to(map)

map
```

```python
# Create data frame with jsut the start_station_id and day_of_week
data_to_plot_2 = station_rec[['Unnamed: 0', 'start_station_latitude', 'start_station_longitude', 'end_station_latitude', 'end_station_longitude', 'trip_duration', 'trip_length']]
data_to_plot_2.head()
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [40.7, -74], zoom_start = 10)

# Choropleth maps bind Pandas Data Frames and json geometries to allow us a quick visualization of data combination
folium.Choropleth(
    geo_data = ny_geo,
    data = data_to_plot_2,
    columns = ['Unnamed: 0', 'trip_duration'],
    key_on = 'feature.properties.name',
    fill_color = 'Pastel2', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "trip_duration").add_to(map)
folium.LayerControl().add_to(map)

map
```
